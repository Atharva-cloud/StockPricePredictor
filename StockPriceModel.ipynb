# RELIANCE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ril = pd.read_csv('C:/Users/athar/OneDrive/Documents/RELIANCE.NS.csv')
```

```python
ril.head()
```

```python
ril.shape
```

```python
ril.describe()
```

```python
ril['Close'].isnull().sum()
```

```python
ril=ril.dropna(axis=0,how='any')
```

```python
ril['Close'].isnull().sum()
```

```python
plt.plot(ril['Close'])
```

```python
ril1 = ril.reset_index()['Close']
```

```python
ril1
```

```python
from sklearn.preprocessing import MinMaxScaler 
sc = MinMaxScaler(feature_range = (0,1))
ril1 = sc.fit_transform(np.array(ril1).reshape(-1,1))
```

```python
ril1
```

```python
training_size_ril = int(len(ril1)*0.70)
test_size_ril =len(ril1)-training_size_ril
train_data_ril, test_data_ril = ril1[0:training_size_ril,:], ril1[training_size_ril:len(ril1),:1]
```

```python
training_size_ril, test_size_ril
```

```python
train_data_ril
```

```python
test_data_ril
```

```python
def create_dataset_ril(dataset, time_step=1):
    dataX_ril, dataY_ril = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
        dataX_ril.append(a)
        dataY_ril.append(dataset[i + time_step, 0])
    return np.array(dataX_ril), np.array(dataY_ril)
```

```python
time_step = 100
X_train_ril, y_train_ril = create_dataset_ril(train_data_ril, time_step)
X_test_ril, ytest_ril = create_dataset_ril(test_data_ril, time_step)
```

```python
print(X_train_ril.shape), print(y_train_ril.shape)
```

```python
print(X_test_ril.shape) ,print(ytest_ril.shape)
```

```python
X_train_ril =X_train_ril.reshape(X_train_ril.shape[0],X_train_ril.shape[1] , 1)
X_test_ril = X_test_ril.reshape(X_test_ril.shape[0],X_test_ril.shape[1] , 1)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
model_ril=Sequential()
model_ril.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model_ril.add(LSTM(50,return_sequences=True))
model_ril.add(LSTM(50))
model_ril.add(Dense(1))
```

```python
model_ril.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model_ril.fit(X_train_ril,y_train_ril,validation_data=(X_test_ril,ytest_ril),epochs=100,batch_size=64,verbose=1)
```

```python
model_ril.save('ril_model.h5')
```

```python
train_predict_ril=model_ril.predict(X_train_ril)
test_predict_ril=model_ril.predict(X_test_ril)
```

```python
train_predict_ril=sc.inverse_transform(train_predict_ril)
test_predict_ril=sc.inverse_transform(test_predict_ril)
```

```python
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_train_ril,train_predict_ril))
```

```python
math.sqrt(mean_squared_error(ytest_ril,test_predict_ril))
```

```python
look_back=100
trainPredictPlot_ril = np.empty_like(ril1)
trainPredictPlot_ril[:, :] = np.nan
trainPredictPlot_ril[look_back:len(train_predict_ril)+look_back, :] = train_predict_ril
# shift test predictions for plotting
testPredictPlot_ril = np.empty_like(ril1)
testPredictPlot_ril[:, :] = np.nan
testPredictPlot_ril[len(train_predict_ril)+(look_back*2)+1:len(ril1)-1, :] = test_predict_ril
# plot baseline and predictions
plt.plot(sc.inverse_transform(ril1))
plt.plot(trainPredictPlot_ril)
plt.plot(testPredictPlot_ril)
plt.show()
```

```python
len(test_data_ril)
```

```python
x_input_ril=test_data_ril[270:].reshape(1,-1)
x_input_ril.shape
```

```python
temp_input_ril=list(x_input_ril)
temp_input_ril=temp_input_ril[0].tolist()
```

```python
temp_input_ril
```

```python
lst_output_ril=[]
n_steps=100
i=0
while(i<30):
    
    if(len(temp_input_ril)>100):
        #print(temp_input_ril)
        x_input_ril=np.array(temp_input_ril[1:])
        print("{} day input {}".format(i,x_input_ril))
        x_input_ril=x_input_ril.reshape(1,-1)
        x_input_ril = x_input_ril.reshape((1, n_steps, 1))
        #print(x_input_ril)
        yhat_ril = model_ril.predict(x_input_ril, verbose=0)
        print("{} day output {}".format(i,yhat_ril))
        temp_input_ril.extend(yhat_ril[0].tolist())
        temp_input_ril=temp_input_ril[1:]
        #print(temp_input_ril)
        lst_output_ril.extend(yhat_ril.tolist())
        i=i+1
    else:
        x_input_ril = x_input_ril.reshape((1, n_steps,1))
        yhat_ril = model_ril.predict(x_input_ril, verbose=0)
        print(yhat_ril[0])
        temp_input_ril.extend(yhat_ril[0].tolist())
        print(len(temp_input_ril))
        lst_output_ril.extend(yhat_ril.tolist())
        i=i+1
    

print(lst_output_ril)
```

```python
day_new_ril=np.arange(1,101)
day_pred_ril=np.arange(101,131)
```

```python
len(ril1)
```

```python
plt.plot(day_new_ril,sc.inverse_transform(ril1[1132:]))
plt.plot(day_pred_ril,sc.inverse_transform(lst_output_ril))
```

```python
ril3=ril1.tolist()
ril3.extend(lst_output_ril)
plt.plot(ril3[1100:])
```
